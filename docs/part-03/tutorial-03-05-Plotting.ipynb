# Section 1 - Creating Basic Plots

The Plots.jl package simplifies the process of creating a wide array of
plots, from simple line plots to complex 3D visualizations. The
following illustrates the basic syntax of a line plot for some random
data. First, we define a variable `x` with a range from 1 to 10 for the
x axis. Then, we define `y` with 10 random numbers. For example:

```julia
x = 1:10      
y = rand(10)
line_plot = plot(x, y, title="Simple Line Plot", xlabel="x axis", ylabel="y axis", legend=false)
display(line_plot)
```

We call `plot()` and use x and y as main arguments. Furthermore, we have
a variety of choices to add further information, such as the title,
labels, a legend, colors, opacity and so on. More on this later.

Add multiple series to a plot. Instead of one series, we can also add
multiple series. This is done, for example, by defining one plot (as
before) and then calling `plot!()` while using the first plot as first
argument in the function `plot!()`.

For example:

```julia
x = 1:20
a = rand(20)
b = rand(20)
multi_plot = plot(x, a, title="Many Series", xlabel="x axis", ylabel="y axis", label="a")
plot!(multi_plot, x, b, label="b")
display(multi_plot)
```

> **Note**
>
> The `!` after `plot` adds to the existing plot.

# Section 3 - Saving Plots to Files

Plots.jl supports saving your plots to various file formats including
PNG, SVG, and PDF, enabling you to use your plots outside of Julia. The
function to save plots is `savefig()`, the first argument is the plot
itself and the second argument is the `path/filename` format as string.
Replace `path` with the path, the `filename` with the actual name and
`format` with the file format, e.g. `pdf`, `png`, … . For example, if
you want to save your file as PDF, you would just name it
`path/filename.pdf`.

# Section 4 - Advanced Plotting Techniques

Let’s explore some other common plot types. While you don’t have to
solve any task here, the code might come in handy later on during the

## Bar Plot

```julia
# Bar plot
x_categories = ["A", "B", "C", "D"]
y_values = [15, 23, 18, 30]
bar_plot = bar(x_categories, y_values, title="Bar Plot Example")
display(bar_plot)
```

## Histogram

```julia
data = randn(1000)
hist_plot = histogram(data, bins=30, title="Histogram Example")
display(hist_plot)
```

## Box Plot

```julia
# Box plot
group = repeat(1:4, inner=50)
y = randn(200) .+ group
box_plot = boxplot(group, y, title="Box Plot Example")
display(box_plot)
```
